```python
# import
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# show seaborn datasets
sns.get_dataset_names()
```

```python
# Dataset iris
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.describe()
```

```python
iris.columns
```

```python
# Histogram with kernel density - Univariate distribution
sns.distplot(iris.sepal_length, rug = True, fit = stats.gausshyper);
```

```python
# Scarterplot - Bivariate distribution
sns.jointplot(x = "sepal_length", y = "petal_length", data = iris);
```

```python
# Scarterplot - Bivariate distribution
with sns.axes_style('white'):
    sns.jointplot(x = "sepal_length", y = "petal_length", data = iris, kind = 'hex', color = 'k')
```

```python
# Scarterplot - Bivariate distribution
sns.pairplot(iris)
```

```python
# Linear Relation
# load dataset tips
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe()
```

```python
# Scarterplot - Regression Linear
sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "reg")
```

```python
# Regression Linear
sns.lmplot(x = "total_bill", y = "tip", data = tips)
```

```python
# Non Linear Regression
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", data = tips)
```

```python
# Non Linear Regression
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", row = "sex", data = tips, size = 4)
```

```python
# Non Linear Regression
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", col = "day", col_wrap = 2, data = tips, size = 4)
```

```python
# Load dataset anscombe
anscombe = sns.load_dataset('anscombe')
```

```python
anscombe.head()
```

```python
anscombe.describe()
```

```python
# Scarterplot - NonLinear Regression
sns.lmplot(x = 'x', y = 'y', data = anscombe.query("dataset == 'II'"), ci = None, scatter_kws = {"s": 80})
```

```python
# Viewing outliers
sns.lmplot(x = 'x', y = 'y', data = anscombe.query("dataset == 'III'"), ci = None, scatter_kws = {"s": 80})
```

```python
# Graphics for Cateorical Variable
sns.stripplot(x = "day", y = "total_bill", data = tips)
```

```python
# barplot
sns.barplot(x = "day", y = "total_bill", hue = "sex", data = tips)
```

```python
# boxplot
sns.boxplot(x = "day", y = "total_bill", hue = "time", data = tips)
```
